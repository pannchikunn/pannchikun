#  11 Browein's formula

```python
def y(n):
    if n < 1: return (2**(1/2)-1)
    return (1-(1- pow(y(n-1), 4))**(1/4)) / (1 + (1 - pow(y(n-1), 4))**(1/4))



def a(n):
    if n < 1 : return 6-4*(2**(1/2))
    return ((1 + y(n))**4)*a(n-1) - (2**(2*n + 1))*y(n)*(1 + y(n) * y(n)**2)
```

```python
numerical_approx(1/a(1), digits=100)
```

```python
numerical_approx(1/a(2), digits=100)
```
