# ロジスティック写像

```python
def logistic(r):
    x_0 = [0.4]
    x = [0.4]
    N = 100
    for n in range(0, N):
        x.append( r*x[n]*(1.0-x[n]))
    plt.ylim(0, 1)
    plt.plot(x, 'ro', x, 'b')
    plt.plot(x_0, 'bo')
```

# リターンマップ

```python
def PlotGraph(func):
    xlist = []
    ylist = []
    for x in np.arange(0,1.0,0.01):
        xlist.append(x)
        ylist.append(func(x))
    plt.axis([0.0, 1.0, 0.0, 1.0])
    plt.plot(xlist, ylist)
```

```python
def Cobweb(func, initial):
    vertices = []
    x = initial
    y = 0
    vertices.append([x,y])
    for n in range(1, 20):
        y = func(x)
        vertices.append([x, y])
        
        x = y
        vertices.append([x, y])
    
    vertices = np.array(vertices)
    plt.plot(vertices[:, 0], vertices[:, 1], 'r--')
```

```python
PlotGraph(lambda x: x)
PlotGraph(lambda x: 0.9 * x * (1- x))
Cobweb(lambda x: 0.9 * x * (1- x), 0.4)
```

```python
PlotGraph(lambda x: x)
PlotGraph(lambda x: 1.5 * x * (1- x))
Cobweb(lambda x: 1.5 * x * (1- x), 0.4)
```

# 分岐

```python
def logistic(x,r):
    y = r*x*(1.0-x)
    return y 


def fillit(n1,n2,x0,r):
    x = x0  # initial x value
    z = np.linspace(0.0,1.0,n2-n1)  
    for i in range(0,n1):   
        x = logistic(x,r)

    for i in range(0,n2-n1):   
        x = logistic(x,r)
        z[i] = x

    return z  



def mkplot(r_min,nr): 
    r_max = 4.0 
    rarr = np.linspace(r_min,r_max,nr)  
    n1=100  
    n2=200  
    x0=0.5  
    for i in range(0,nr):  
        r = rarr[i]
        y=fillit(n1,n2,x0,r) 
        x=y*0.0 + r   
        plt.plot(x,y,'ko',markersize=1)  



plt.figure(figsize = (10,12))
plt.xlabel('r',fontsize=20)
plt.ylabel('$X_n$',fontsize=20, rotation =0)
plt.axis([0, 4.0, 0, 1.0])
# this makes the plot!
mkplot(0,1000)
```
